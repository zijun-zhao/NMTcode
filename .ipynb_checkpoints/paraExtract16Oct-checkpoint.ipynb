```python
from scipy.spatial import distance
import ast
#ns.features['soma_features'].raw_data
def metric_Cal(raw_data, regionINFO):
    somaDF = raw_data.copy()
    for idx in somaDF.index.tolist():
        tempCoor = np.array([(somaDF.loc[idx,'x'],somaDF.loc[idx,'y'],somaDF.loc[idx,'z'])])
        zs = np.sum(distance.cdist(tempCoor , ast.literal_eval(regionINFO.loc[idx,'Coor']), 'sqeuclidean'))
        somaDF.loc[idx,'SUMsqeuclidean'] = zs
    return somaDF.copy()
```

```python
raw_data =ns.features['soma_features'].raw_data
somaDF = raw_data.copy()
scaled_somaDF = pd.DataFrame()
scaled_somaDF['x']=(somaDF['x']/nmt.annotation.space["x"]).astype(int)
scaled_somaDF['y']=(somaDF['y']/nmt.annotation.space["y"]).astype(int)
scaled_somaDF['z']=(somaDF['z']/nmt.annotation.space["z"]).astype(int)
for idx in somaDF.index.tolist():
    tempCoor = np.array([(scaled_somaDF.loc[idx,'x'],scaled_somaDF.loc[idx,'y'],scaled_somaDF.loc[idx,'z'])])
    tempID = nmt.annotation.array[(scaled_somaDF.loc[idx,'x'],scaled_somaDF.loc[idx,'y'],scaled_somaDF.loc[idx,'z'])]
    zs = np.sum(distance.cdist(tempCoor , ast.literal_eval(regionINFO.loc[tempID,'Coor']), 'sqeuclidean'))
    somaDF.loc[idx,'SUMsqeuclidean'] = zs
```
