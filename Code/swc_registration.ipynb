```python
import os
path = '/home/penglab/Desktop/Warp_swc_Step_test'
for brain_id in os.listdir(path ):
    if not (brain_id) in list(map(str, brain_regis.index.tolist())):
        print("Skip file " + str(brain_id.split('/')[-1]) )
        continue

    folder_affine =  os.path.join(os.path.join(path, subpath),'affine')
    exist_affine = os.path.exists(folder_affine)
    if not exist_affine:      
        print('Creat new folder'+ str(folder_affine))
        os.makedirs(folder_affine)  

    folder_ori =  os.path.join(os.path.join(path, subpath),'ori')
    exist_ori = os.path.exists(folder_ori)
    if not exist_ori:         
        print('Creat new folder '+ str(folder_ori))
        os.makedirs(folder_ori)  

    folder_resample =  os.path.join(os.path.join(path, subpath),'resample')
    exist_resample = os.path.exists(folder_resample )
    if not exist_resample:   
        print('Creat new folder '+ str(folder_resample))
        os.makedirs(folder_resample)  

    folder_stps =  os.path.join(os.path.join(path, subpath),'stps')
    exist_stps = os.path.exists(folder_stps)
    if not exist_stps:   
        print('Creat new folder '+ str(folder_stps))
        os.makedirs(folder_stps)  
    x_ds_size = brain_regis.loc[int(brain_id),'x_downsample']
    y_ds_size = brain_regis.loc[int(brain_id),'y_downsample']
    z_ds_size = brain_regis.loc[int(brain_id),'z_downsample']
```

```python
    for iter_swc in os.listdir(folder_ori):
        if not iter_swc.endswith(("swc", "SWC")):
            continue
        (neuron_info,extension) = os.path.splitext(iter_swc)
        temp_brain = int(neuron_info.split('_')[0])
        print('Neuron information: ' + str(neuron_info))
        #print(temp_brain)
        path_old  = os.path.join(folder_ori, iter_swc)
        path_new  = os.path.join(folder_resample, neuron_info)+ '.' + 'x' + str(round(x_ds_size,2)) + 'y' + str(round(y_ds_size,2)) + 'z'+ str(round(z_ds_size,2)) + str(extension)
        n_skip = 0
        with open(path_old, "r") as f:
            for line in f.readlines():
                line = line.strip()
                if line.startswith("#"):
                    n_skip += 1
                else:
                    break
        names = ["##n", "type", "x", "y", "z", "r", "parent"]
        swc = pd.read_csv(path_old, index_col=0, skiprows=n_skip, sep=" ",
                  usecols=[0, 1, 2, 3, 4, 5, 6],
                  names=names
                  )
        swc.loc[:,'x'] = swc.loc[:,'x']/round(x_ds_size,1)
        swc.loc[:,'y'] = swc.loc[:,'y']/round(y_ds_size,1)
        swc.loc[:,'z'] = swc.loc[:,'z']/round(z_ds_size,1)
        assert(max( swc.loc[:,'x'])) < brain_regis.loc[int(brain_id),'x_after'], " x has exceeds the boundary"
        assert(max( swc.loc[:,'y'])) < brain_regis.loc[int(brain_id),'y_after'], " y has exceeds the boundary"
        assert(max( swc.loc[:,'z'])) < brain_regis.loc[int(brain_id),'z_after'], " z has exceeds the boundary"
        swc.to_csv(path_new, sep=" ")
        print('Saving current .swc to ' + str(path_new) + '\n')
```
