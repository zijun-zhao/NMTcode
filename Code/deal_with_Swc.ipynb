```python
import os
import shutil
path = '/home/penglab/Documents/result/resampled_ALL'

for i_swc in os.listdir(path):
    if i_swc.find(".swc.resampled.swc") > 0:
        (neuronname,extension) = os.path.splitext(i_swc)
        (neuronname,extension) = os.path.splitext(neuronname)
        (neuronname,extension) = os.path.splitext(neuronname)
        (neuronname,extension) = os.path.splitext(neuronname)
        os.rename(os.path.join(path, i_swc), os.path.join(path, i_swc.replace(".swc.resampled.swc", ".swc")))
```

```python
import pandas as pd
folder_resample = '/home/penglab/Documents/result/resampled_ALL_fi'
for iter_swc in os.listdir(path):
    if not iter_swc.endswith(("swc", "SWC")):
        continue
    (neuron_info, extension) = os.path.splitext(iter_swc)
    print('Neuron information: ' + str(neuron_info))
    path_old = os.path.join(path, iter_swc)
    path_new = os.path.join(folder_resample, neuron_info) + '.'  + str(extension)
    n_skip = 0
    with open(path_old, "r") as f:
        for line in f.readlines():
            line = line.strip()
            if line.startswith("#"):
                n_skip += 1
            else:
                break
    names = ["##n", "type", "x", "y", "z", "r", "parent"]
    swc = pd.read_csv(path_old, index_col=0, skiprows=n_skip, sep=" ",
                      usecols=[0, 1, 2, 3, 4, 5, 6],
                      names=names
                      )
    

    swc.to_csv(path_new, sep=" ")

    print('Saving current file to ' + str(path_new))
```
