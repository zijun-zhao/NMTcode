### Copy file from jbasumOCT to ww2scale, here the previous version's file name containing multiple dot

```python
import os
import shutil
for i_swc in os.listdir('/home/penglab/Desktop/jbasumOCT' ):
    folder = os.path.exists('/home/penglab/FeaCal/ww2scale')
    if not folder:         
        os.makedirs('/home/penglab/FeaCal/ww2scale')       
    oldname = '/home/penglab/Desktop/jbasumOCT' + '/'+ str(i_swc)
    (filename,extension) = os.path.splitext(i_swc)
    (filename,extension) = os.path.splitext(filename)
    (filename,extension) = os.path.splitext(filename)
    (filename,extension) = os.path.splitext(filename)
    newname = '/home/penglab/FeaCal/ww2scale' + '/'+ str(filename)+'.swc'
    shutil.copyfile(oldname,newname)
```

###  For all items in the ww2scale, scale it and move to scaledWW

```python
swc_path = '/home/penglab/FeaCal/ww2scale'
from neuro_morpho_toolbox.neuron_set import load_swc_list
#Read the swc inside this path to adictionary
dicAll = load_swc_list(swc_path)
for i in dicAll.keys():
    dicAll[i].swc.loc[:,'x'] = dicAll[i].swc.loc[:,'x'].copy()-20
    dicAll[i].swc.loc[:,'x'] = dicAll[i].swc.loc[:,'x'].copy()*25
    dicAll[i].swc.loc[:,'y'] = dicAll[i].swc.loc[:,'y'].copy()*25
    dicAll[i].swc.loc[:,'z'] = dicAll[i].swc.loc[:,'z'].copy()*25
    pathdes = '/home/penglab/FeaCal/scaledWW/'
    dicAll[i].swc.to_csv(pathdes+str(i)+'.swc', sep=" ")
    print(i)
```

###  Count number of files under scaledWW
